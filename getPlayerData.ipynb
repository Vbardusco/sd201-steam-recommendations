```python
import json

with open('SteamGameData.json', 'r') as f:
    dataset = f.read()
    dataset = json.loads(dataset)
    print(len(dataset))

print(dataset[0])
```

```python
def getPlayerData(steamId, dataset):
  playerInfo = []
  for line in dataset:
    if line["steamid"]==steamId:
      playerInfo.append(line)
  return json.dumps(playerInfo)
```
